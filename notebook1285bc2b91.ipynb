```python
import numpy as np
n=len(df)
count_class_0=len(df[df['Class']==0])
count_class_1=len(df[df['Class']==1])

print("Dataset percent for class 0 : ",np.round(count_class_0*100/n , 3) , "%  , Dataset count for class 0 : " , count_class_0)
print("Dataset percent for class 1 : ",np.round(count_class_1*100/n ,3) , "%  , Dataset count for class 0 : "  , count_class_1)
```
